# Calculate DQN Statistics

```python
import csv
import numpy as np

# Read the data from episode_metrics.csv
episode_data = []
with open("DQN_episode_metrics.csv", mode="r") as file:
    reader = csv.DictReader(file)
    for row in reader:
        episode_data.append({
            "collision": int(row["Collision"]),
            "average_speed": float(row["Average Speed"]),
            "right_lane_percentage": float(row["Right Lane Percentage"]),
            "lane_changes": int(row["Lane Changes"]),
            "unsafe_lane_changes": int(row["Unsafe Lane Changes"]),
            "average_front_car_distance": float(row["Average Front Car Distance"]),
        })

# Calculate statistics
num_episodes = len(episode_data)
total_collisions = sum(d["collision"] for d in episode_data)
average_speed = np.mean([d["average_speed"] for d in episode_data])
average_right_lane_percentage = np.mean([d["right_lane_percentage"] for d in episode_data])
average_lane_changes = np.mean([d["lane_changes"] for d in episode_data])
total_lane_changes = sum(d["lane_changes"] for d in episode_data)
average_unsafe_lane_changes = np.mean([d["unsafe_lane_changes"] for d in episode_data])
unsafe_lane_change_percentage = (sum(d["unsafe_lane_changes"] for d in episode_data) / total_lane_changes) * 100 if total_lane_changes > 0 else 0
average_front_car_distance = np.mean([d["average_front_car_distance"] for d in episode_data])

collision_rate = (total_collisions / num_episodes) * 100

# Prepare statistics summary
statistics_summary = {
    "Total Episodes": num_episodes,
    "Collision Rate (%)": collision_rate,
    "Average Speed (m/s)": average_speed,
    "Average Right Lane Percentage (%)": average_right_lane_percentage,
    "Average Lane Changes": average_lane_changes,
    "Average Unsafe Lane Changes": average_unsafe_lane_changes,
    "Unsafe Lane Change Percentage (%)": unsafe_lane_change_percentage,
    "Average Front Car Distance (m)": average_front_car_distance
}

# Print statistics summary
print("Statistics Summary:")
for key, value in statistics_summary.items():
    print(f"{key}: {value}")

```

# Draw box plots

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Prepare data for box plots
metrics_data = {
    'Speed (m/s)': [d['average_speed'] for d in episode_data],
    'Right Lane (%)': [d['right_lane_percentage'] for d in episode_data],
    'Lane Changes': [d['lane_changes'] for d in episode_data],
    'Unsafe Lane Changes': [d['unsafe_lane_changes'] for d in episode_data],
    'Front Car Distance (m)': [d['average_front_car_distance'] for d in episode_data]
}



# Optional: Create individual box plots if the scales are too different
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes = axes.ravel()

for idx, (metric, values) in enumerate(metrics_data.items()):
    if idx < len(axes):
        sns.boxplot(y=values, ax=axes[idx])
        axes[idx].set_title(metric)
        axes[idx].set_ylabel('Value')

# Remove extra subplot if any
if len(metrics_data) < len(axes):
    fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()
```

## Comparison

```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data from both CSV files
def load_data(filename):
    data = []
    with open(filename, mode="r") as file:
        reader = csv.DictReader(file)
        for row in reader:
            data.append({
                "collision": int(row["Collision"]),
                "average_speed": float(row["Average Speed"]),
                "right_lane_percentage": float(row["Right Lane Percentage"]),
                "lane_changes": int(row["Lane Changes"]),
                "unsafe_lane_changes": int(row["Unsafe Lane Changes"]),
                "average_front_car_distance": float(row["Average Front Car Distance"]),
            })
    return data

# Load DQN and DDQN data
dqn_data = load_data("DQN_episode_metrics.csv")
ddqn_data = load_data("DDQN_episode_metrics.csv")

# Function to calculate metrics
def calculate_metrics(data):
    metrics = {
        "Collision Rate (%)": np.mean([d['collision'] for d in data]) * 100,
        "Average Speed (m/s)": np.mean([d['average_speed'] for d in data]),
        "Right Lane Percentage (%)": np.mean([d['right_lane_percentage'] for d in data]),
        "Lane Changes": np.mean([d['lane_changes'] for d in data]),
        "Unsafe Lane Change Percentage (%)": (np.mean([d['unsafe_lane_changes'] for d in data]) / np.mean([d['lane_changes'] for d in data])) * 100,
        "Front Car Distance (m)": np.mean([d['average_front_car_distance'] for d in data])
    }
    return metrics

# Calculate metrics for DQN and DDQN
dqn_metrics = calculate_metrics(dqn_data)
ddqn_metrics = calculate_metrics(ddqn_data)

# Convert to DataFrame for easier comparison
metrics_df = pd.DataFrame([dqn_metrics, ddqn_metrics], index=["DQN", "DDQN"]).T
print(metrics_df)



```

```python
# Plotting metrics comparison
metrics_df = metrics_df.reset_index().rename(columns={'index': 'Metric'})
plt.figure(figsize=(12, 8))
sns.barplot(x="Metric", y="value", hue="variable", data=pd.melt(metrics_df, id_vars=['Metric']), palette="Set2")
plt.title("Comparison of DQN and DDQN Metrics")
plt.ylabel("Value")
plt.xlabel("Metric")
plt.xticks(rotation=45)
plt.legend(title="Algorithm")
plt.tight_layout()
plt.show()
```
